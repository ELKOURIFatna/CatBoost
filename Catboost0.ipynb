```python
from catboost import CatBoostClassifier
```

```python
import pandas as pd
```

```python
#data path 
file_path = 'D:\Python\CatBoost\heart_disease_uci.csv'

#read data with pandas from excel file
data = pd.read_csv(file_path)

#show first 5 lignes 
data.head()
```

```python
data.drop(columns = ['dataset'], inplace= True)
data.drop(columns = ['id'], inplace= True)
```

```python
# List of categorical features
categorical_features = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'thal']
```

```python
#filling th NAN with a string so can catboost handle it correctly
data['sex'] = data['sex'].fillna('Missing')
data['cp'] = data['cp'].fillna('Missing')
data['fbs'] = data['fbs'].fillna('Missing')
data['restecg'] = data['restecg'].fillna('Missing')
data['exang'] = data['exang'].fillna('Missing')
data['slope'] = data['slope'].fillna('Missing')
data['thal'] = data['thal'].fillna('Missing')
```

```python
data.info()
```

```python
data.describe()
```

```python
#to see which data values are null and missing values
print(data.isnull().sum())
```

```python
#grant data to variable
features = data.drop('num', axis=1)

#target variable
target = data['num']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
model = CatBoostClassifier(iterations=1000, learning_rate=0.1, depth=6, verbose=10)

#getting their indices
categorical_features_indices = [features.columns.get_loc(col) for col in categorical_features]

#verbose=10  # Print progress every 10 iterations
# Train the model
model.fit(X_train, y_train, cat_features = categorical_features, eval_set=(X_test, y_test))
```

```python
#model predict
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
