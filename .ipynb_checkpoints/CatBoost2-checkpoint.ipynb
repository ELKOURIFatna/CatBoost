```python
from catboost import CatBoostClassifier
import pandas as pd
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score, roc_curve, precision_recall_curve, auc
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import numpy as np
```

```python
# Data path
file_path = 'D:\Python\CatBoost\heart_disease_uci.csv'

# Read data with pandas from CSV file
data = pd.read_csv(file_path)
```

```python
# Show first 5 rows
print(data.head())
```

```python
# Check for missing values
print(data.isnull().sum())
```

```python
# Remove unnecessary columns
data.drop(columns=['dataset', 'id'], inplace=True)

# Separate features and target
features = data.drop('num', axis=1)
target = data['num']
```

```python
categorical_features = ['sex', 'cp', 'restecg', 'slope', 'thal']
for category in categorical_features:
    print(features[category].unique())
```

```python
# Fill categorical NaNs with a string so CatBoost can handle it correctly
categorical_features = ['sex', 'cp', 'restecg', 'slope', 'thal']
for col in categorical_features:
    features[col] = features[col].fillna('Missing')

# Fill numeric NaNs with the mean
numeric_features = features.select_dtypes(include=['float64', 'int64']).columns
for col in numeric_features:
    features[col] = features[col].fillna(features[col].mean())
```

```python
# Check for missing values again
print(features.isnull().sum())
```

```python
# Fill true/false NaNs
true_false_features = ['fbs', 'exang']
for var in true_false_features:
    mode_value = features[var].mode()[0]
    features[var] = features[var].fillna(mode_value)
```

```python
# Check for missing values again
print(features.isnull().sum())
```

```python
all_features = categorical_features + true_false_features
print(all_features)
```

```python
# Encode categorical columns using OneHotEncoder
encoder = OneHotEncoder(drop='first', sparse_output=False)
encoded_categorical = encoder.fit_transform(features[all_features])
encoded_categorical_df = pd.DataFrame(encoded_categorical, columns=encoder.get_feature_names_out(all_features))
```

```python
encoded_categorical_df
```

```python
# Drop original categorical columns and concatenate encoded ones
features = features.drop(all_features, axis=1)
features = pd.concat([features, encoded_categorical_df], axis=1)
```

```python
features
```

```python
# Convert boolean columns to strings (if any)
#for col in features.columns:
#    if features[col].dtype == 'bool':
#        features[col] = features[col].astype(str)

# Get indices of categorical features
#categorical_features_indices = [features.columns.get_loc(col) for col in categorical_features]
```

```python
# Split data into train (80%) and test (20%)
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
# Apply SMOTE to handle imbalanced data
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)
```

```python
# Scale numeric features (optional for CatBoost, but can help)
scaler = StandardScaler()
X_resampled[numeric_features] = scaler.fit_transform(X_resampled[numeric_features])
X_test[numeric_features] = scaler.transform(X_test[numeric_features])
```

```python
# Hyperparameter tuning using GridSearchCV
#param_grid = {
#    'iterations': [1000, 2000],
#    'learning_rate': [0.01, 0.05, 0.1],
#    'depth': [4, 6, 8],
#    'l2_leaf_reg': [1, 3, 5]
#}
```

```python
# Best model from grid search
#best_model = grid_search.best_estimator_
```

```python
model = CatBoostClassifier(iterations=1000, learning_rate=0.05, depth=6, verbose=100)
#verbose=100  # Print progress every 100 iterations
```

```python
# Cross-validation to ensure generalization
cv_scores = cross_val_score(model, X_resampled, y_resampled, cv=5, scoring='accuracy')
print(f"Cross-Validation Accuracy: {np.mean(cv_scores) * 100:.2f}%")
```

```python
# Train the best model
model.fit(X_resampled, y_resampled, eval_set=(X_test, y_test))
```

```python
# Model predict
y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)
```

```python
y_pred
```

```python
# Assuming y_test is your true labels and y_pred_proba is the predicted probabilities
roc_auc = roc_auc_score(y_test, y_pred_proba, multi_class='ovr')  # or 'ovo'
```

```python
# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred) * 100, "%")
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("ROC-AUC Score:", roc_auc)
```

```python
# Plot feature importance
feature_importance = model.get_feature_importance()
plt.barh(features.columns, feature_importance)
plt.xlabel('Feature Importance')
plt.ylabel('Features')
plt.title('Feature Importance Plot')
plt.show()
```

```python
from sklearn.preprocessing import label_binarize

# Binarize the true labels (one-hot encoding)
y_test_bin = label_binarize(y_test, classes=[0, 1, 2, 3, 4])  # Adjust classes based on your data

# Compute ROC curve and ROC area for each class
n_classes = y_test_bin.shape[1]
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_pred_proba[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plot ROC curves for each class
plt.figure()
colors = ['blue', 'red', 'green', 'orange', 'purple']  # Adjust colors based on the number of classes
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2,
             label='ROC curve of class {0} (AUC = {1:0.2f})'.format(i, roc_auc[i]))

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Multi-Class ROC Curve')
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve, auc
from sklearn.preprocessing import label_binarize
import matplotlib.pyplot as plt

# Binarize the true labels (one-hot encoding)
y_test_bin = label_binarize(y_test, classes=[0, 1, 2, 3, 4])  # Adjust classes based on your data

# Compute Precision-Recall curve and AUC for each class
n_classes = y_test_bin.shape[1]
precision = dict()
recall = dict()
pr_auc = dict()

for i in range(n_classes):
    precision[i], recall[i], _ = precision_recall_curve(y_test_bin[:, i], y_pred_proba[:, i])
    pr_auc[i] = auc(recall[i], precision[i])

# Plot Precision-Recall curves for each class
plt.figure()
colors = ['blue', 'red', 'green', 'orange', 'purple']  # Adjust colors based on the number of classes
for i, color in zip(range(n_classes), colors):
    plt.plot(recall[i], precision[i], color=color, lw=2,
             label='Precision-Recall curve of class {0} (AUC = {1:0.2f})'.format(i, pr_auc[i]))

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Multi-Class Precision-Recall Curve')
plt.legend(loc="lower left")
plt.show()
```
